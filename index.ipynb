## Data Cleaning
### Loading Data

We begin by importing the necessary library (pandas) and loading the aviation dataset into a DataFrame. This step allows us to inspect and manipulate the data efficiently.

```python
import pandas as pd
df = pd.read_csv("Aviation_Data.csv")
df
```

### Data Inspection

Before cleaning, it’s important to understand the dataset:



```python
df.shape   #shows the number of rows and columns.

df.info()   #provides information about column types and non-null counts.

df.isna().sum().sort_values(ascending=False)  #helps identify columns with missing values.

df.columns #lists all column names for easier selection.
```

### Selecting Relevant Columns

We narrow down the dataset to only the columns relevant for our analysis. 



```python
cols = [
    "Make",
    "Model",
    "Number.of.Engines",
    'Purpose.of.flight',
    "Weather.Condition",
    "Event.Date",
    "Total.Fatal.Injuries",
    "Total.Serious.Injuries",
    "Total.Minor.Injuries",
    "Total.Uninjured"
]

df = df[cols]
df.isna().sum()
```

### Standardizing Data

Text data often contains inconsistencies such as extra spaces or irregular capitalization. We standardize important text columns by stripping whitespace and converting them to title case.


```python
text_cols = ['Make', 'Model','Purpose.of.flight',"Weather.Condition"]

for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

### Handling Null Values

* For injury-related columns, missing values are replaced with 0 because a null entry likely indicates no injuries reported.

* Unknown placeholders such as "UNK", "Unknown", "N/A", or empty strings are converted to pandas.NA to mark them as missing.

* Rows missing critical aircraft information (Make, Model, Weather.Condition) are dropped because they are essential for analysis.



```python
injury_cols = [
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries',
    'Total.Uninjured'
]

df[injury_cols] = df[injury_cols].fillna(0)
df = df.replace(["UNK","Unk","Unknown", "N/A", ""], pd.NA)
df = df.dropna(subset=['Make','Model',"Weather.Condition",'Number.of.Engines'])
df.info()
```

## Data Manipulation
### Creating a Severity Factor

We define a severity score to quantify the overall impact of an incident:

* Fatal injuries are weighted 5 times.

* Serious injuries are weighted 3 times.

* Minor injuries are weighted 1 time.

This allows us to rank incidents based on severity.


```python
df['Severity.Score'] = (
    5 * df['Total.Fatal.Injuries'] +
    3 * df['Total.Serious.Injuries'] +
    1 * df['Total.Minor.Injuries']
)
```

### Date Processing

The Event.Date column is converted to a proper datetime format, and a new column Year is created to analyze trends over time.


```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Year'] = df['Event.Date'].dt.year   #converts date to year
df = df[df['Year'] >= 1985]    #keeps only recent data from 1985 onwards to ensure relevance.
```

### Sorting According to Frequency

To simplify analysis, we focus on the most common makes, models, engines, and flight purposes:

* Top 5 aircraft makes

* Top 10 models

* Top 5 engine configurations




```python

top_models = df['Model'].value_counts().head(10).index


df_top = df[df['Model'].isin(top_models)]

# Get top 10 most common flight purposes
top_purposes = df['Purpose.of.flight'].value_counts().head(10).index

# Group all other purposes as 'Other'
df_top['Purpose_Grouped'] = df_top['Purpose.of.flight'].where(
    df['Purpose.of.flight'].isin(top_purposes),
    'Other'
)


df_top = df_top[~((df_top['Make'] == 'Piper') & (df_top['Model'] == '180'))]  #Remove unwanted row with misplaced entry
df_top.info()
```

### Saving Cleaned Data

The cleaned and filtered dataset is saved as a new CSV file for further analysis.



```python

df_top.to_csv("aviation_cleaned.csv", index=False)
df_top.isna().sum()
```
